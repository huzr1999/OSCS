```python
import pandas as pd
import argparse
import os

def sample_csv(input_path, output_path, sample_ratio=0.1, random_seed=42):
    # 读取 CSV 文件
    df = pd.read_csv(input_path, index_col=0)  # 忽略第一列索引
    print(f"原始数据集大小: {len(df)}")

    # 按比例随机抽样（不重复）
    sampled_df = df.sample(frac=sample_ratio, random_state=random_seed)
    print(f"抽取样本数: {len(sampled_df)}")

    # 保存到新的 CSV 文件
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    sampled_df.to_csv(output_path, index=True)
    print(f"已保存到: {output_path}")
```

```python
sample_csv("./train-clean-total.csv", "./train-clean.csv", sample_ratio=0.1, random_seed=42)
sample_csv("./dev-clean-total.csv", "./dev-clean.csv", sample_ratio=0.2, random_seed=42)
sample_csv("./test-clean-total.csv", "./test-clean.csv", sample_ratio=0.2, random_seed=42)
```
